```python
info = ['id', 'area', 'size', 'sex', 'birth', 'age']

survey = data2.loc[:, 'a13':'a1610'].columns.tolist()

mental_health = data2[info + survey]
```

```python
mental_health.head()
```

```python
n13_mapping = {
    "매우 불만족한다": 0,
    "약간 불만족한다": 1,
    "보통이다": 2,
    "약간 만족한다": 3,
    "매우 만족한다" : 4
}

mental_health['a13'] = mental_health['a13'].replace(n13_mapping)
```

```python
n14_mapping = {
    "없음": 0,
    "2~6일": 1,
    "7일 이상": 2,
    "거의 매일": 3
}

n14_cols = mental_health.loc[:, 'a141':'a149'].columns.tolist()
mental_health[n14_cols] = mental_health[n14_cols].replace(n14_mapping)
```

```python
n15_mapping = {
    "전혀 방해 받지 않았다" : 0,
    "2주 중 절반 이상은 방해 받았다" : 1,
    "며칠 동안방해 받았다" : 2,
    "거의 매일 방해 받았다" : 3
}

n15_cols = mental_health.loc[:, 'a151':'a157'].columns.tolist()
mental_health[n15_cols] = mental_health[n15_cols].replace(n15_mapping)
```

```python
n16_mapping = {
    "전혀 없음" : 0, 
    "거의 없음" : 1,
    "때때로 있음" : 2,
    "자주 있음" : 3,
    "매우 자주" : 4
}

n16_cols = mental_health.loc[:, 'a1601':'a1610'].columns.tolist()
mental_health[n16_cols] = mental_health[n16_cols].replace(n16_mapping)
```
