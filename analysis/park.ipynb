3. 공원 시설물 현황

```python
facilities = pd.read_excel(
    "04. 2024년 공원현황 2권 공원시설물현황.xlsx",
    sheet_name="시직영(사업소)", 
    header=[0, 1, 2],
    engine="openpyxl",
    skiprows=2,
)

facilities.head()
```

```python
facilities = facilities.iloc[4:]
facilities.head()
```

```python
# 공원명 중복 열 삭제
columns_to_drop_by_index = [37, 73, 102]

facilities = facilities.drop(facilities.columns[columns_to_drop_by_index], axis=1)
facilities.head()
```

```python
# 관리시설 ~ 삭제
facilities = facilities.drop(columns=['관   리   시   설', '적  법  건  물', '부  적  격  건  물'], level=0)
facilities.head()
```

```python
# 시설 인덱스 삭제
facilities.columns = facilities.columns.droplevel(0)
facilities.head()
```

```python
# 기타 행 삭제
mask = facilities.columns.get_level_values(0) != '기   타'

facilities = facilities.loc[:, mask]
facilities.head()
```

```python
# 	Unnamed: 0_level_2 정리
columns_to_remove = ['㎡', '건축면적(㎡)', '면적(㎡)', '㎞', '건축면적', '기', '타석', '주차대수', '면수']

mask = ~facilities.columns.get_level_values(1).isin(columns_to_remove)

facilities = facilities.loc[:, mask]

facilities.columns = facilities.columns.droplevel(1)

facilities.head()
```

```python
# Unnamed: 0_level_1 이름 변경
facilities = facilities.rename(columns={
    'Unnamed: 0_level_1' : '공원명',
    })
facilities.head()
```

```python
facilities.columns = facilities.columns.str.replace('\n', '')
facilities.columns = facilities.columns.str.replace(' ', '')
facilities['공원명'] = facilities['공원명'].str.replace(' ', '')
```

```python
import re

facilities.columns = facilities.columns.str.replace(r'\(.*\)', '', regex=True)
facilities.head()
```

```python
facilities = facilities.fillna(0)
```
